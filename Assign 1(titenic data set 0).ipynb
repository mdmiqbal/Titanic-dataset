```python
#Total no of passengers & survivors
```

```python
total_passengers = len(df)
total_survivors = len(df[df.Survived==1])
total_nonsurvivors = total_passengers - total_survivors
print("Total passenger",total_passengers ,"Total surv"
      ,total_survivors ,
      "Total nonsur",total_nonsurvivors)
```

```python
#total survivor male and female
table = pd.crosstab(df['Survived'],df['Sex'])
print (table)
```

```python
#survivor according to class 
```

```python
df.groupby(['Sex']).size().plot(kind='bar',stacked=True)
plt.title("Ratio of male and female on titanic",fontsize=14)
plt.ylabel('total no of persion')
plt.xlabel('sex');
```

```python
df.groupby(['Pclass']).size().plot(kind='bar',stacked=True)
plt.title("survivour according to class",fontsize=14)
plt.ylabel('survivour')
plt.xlabel('Class');
```

```python
#age group that has most survivour chance
```

```python
age_labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
df['age_group'] = pd.cut(df.Age, range(0, 81, 10), right=False, labels=age_labels)
```

```python

df.groupby(['age_group']).size().plot(kind='bar',stacked=True)
plt.title("Distribution of Age Groups",fontsize=14)
plt.ylabel('Count')
plt.xlabel('Age Group');
```
